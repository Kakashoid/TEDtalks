```python
import re

bd['title'] = bd['title'].str.lower()

bd['title'] = bd['title'].apply(lambda x: re.sub('http\S*', '', x))

bd['title'] = bd['title'].apply(lambda x: re.sub('\s+', ' ', x))

bd['title'] = bd['title'].apply(lambda x: re.sub('[^\w\s]', '', x))
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

bd['title'] = bd['title'].apply(lambda text: " ".join(word for word in text.split() if word not in stop))
```

```python
from wordcloud import WordCloud

wc = WordCloud(background_color="white", max_words=2000, width=800, height=400)

import matplotlib.pyplot as plt
%matplotlib inline

wc.generate(' '.join(bd['title']))

plt.figure(figsize=(24, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
